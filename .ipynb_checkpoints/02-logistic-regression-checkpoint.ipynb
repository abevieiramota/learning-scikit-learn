```python
from sklearn.linear_model import LogisticRegression
import numpy as np
from matplotlib import pyplot as pl
from itertools import product
%matplotlib inline
%run util.py
```

```python
classes = [1, 2]
proporcao = [80, 20]
colors = ['r', 'b']
penalties = ['l1', 'l2']
Cs = [.001, .1, 1., 100]
params = product(penalties, Cs)
total = len(penalties)*len(Cs)

# dados

# [-5, 5] size 100
X0 = np.linspace(-5, 5, 100).reshape((100, 1))
# normal média 0. desvio padão 10. size 100
X1 = np.linspace(0, 10, 100).reshape((100, 1)) + np.random.normal(0., 2., 100).reshape((100, 1))
# 'empilha' horizontalmente X0 e X1
X = np.hstack((X0, X1))
# gera um array de valores em classes com size 100
Y = np.ones((100,))*classes[0]
Y[50:] = classes[1]
# índices randômicos na proporção 80 20, treino e teste, respectivamente
i_treino, i_teste = indices(proporcao)
# partição de X e Y em training set e test set
X_treino, X_teste = X[i_treino, :], X[i_teste, :]
Y_treino, Y_teste = Y[i_treino], Y[i_teste]

fig = pl.figure(figsize=(20, total*4))

for i, (penalty, C) in enumerate(params):
    
    model = LogisticRegression(penalty=penalty, C=C)
    model.fit(X_treino, Y_treino)
    # predição para os dados de teste
    _Y = model.predict(X_teste)
    # predição para os dados de treino
    Y_ = model.predict(X_treino)
    
    
    # dados
    ax0 = fig.add_subplot(total, 5, i*5 + 1)
    ax0.set_title(u'dados')
    # classificação correta - teste
    ax1 = fig.add_subplot(total, 5, i*5 + 2)
    ax1.set_title(u'teste real')
    # classificação do modelo - teste
    ax2 = fig.add_subplot(total, 5, i*5 + 3)
    ax2.set_title(u'penalty %s C %.3f' % (penalty, C))
    # classificação correta - treino
    ax3 = fig.add_subplot(total, 5, i*5 + 4)
    ax3.set_title(u'treino real')
    # classificação do odelo - treino
    ax4 = fig.add_subplot(total, 5, i*5 + 5)
    ax4.set_title(u'penalty %s C %.3f' % (penalty, C))

    for classe, c in zip(classes, colors):
        ax0.scatter(X[Y==classe][:, 0], X[Y==classe][:, 1], c=c)
        ax1.scatter(X_teste[Y_teste==classe][:, 0], X_teste[Y_teste==classe][:, 1], c=c)
        ax2.scatter(X_teste[_Y==classe][:, 0], X_teste[_Y==classe][:, 1], c=c)
        ax3.scatter(X_treino[Y_treino==classe][:, 0], X_treino[Y_treino==classe][:, 1], c=c)
        ax4.scatter(X_treino[Y_==classe][:, 0], X_treino[Y_==classe][:, 1], c=c)
```
