```python
from sklearn.linear_model import LinearRegression
import numpy as np
from matplotlib import pyplot as pl
%matplotlib inline
```

```python
# start: 0 end: 10 size=100
x = np.linspace(0, 10, 100).reshape((100, 1))
# f: x -> x**2
# y = f(x)
y = x*2
# configurações dos ruídos (média, desvio padrão)
noises = [(0., .5), (0., 2.), (0., 4.), (.2, .5), (.2, 2.), (.2, 4.)]
# número de ruídos, para calcular o número de linhas no plot e o tamanho da figure
n = len(noises)

#figure que conterá os plots
fig = pl.figure(figsize=(12, n*4))

# enumerate percorre noises retornando um índice e um elemento por vez
for i, noise in enumerate(noises):
    
    mu = noise[0]
    sigma = noise[1]

    # ruído
    noise = np.random.normal(mu, sigma, 100).reshape((100, 1))

    # y com ruído
    y_ = y + noise

    # sem noise
    ax1 = fig.add_subplot(n, 3, i*3 + 1)
    ax1.scatter(x, y)
    ax1.set_title("sem noise")

    # com noise
    ax2 = fig.add_subplot(n, 3, i*3 + 2)
    ax2.scatter(x, y_)
    ax2.set_title("noise: $\mu=%.2f$ $\sigma=%.2f$" % (mu, sigma))

    # linear model
    # :fit_intercept -> se utiliza ou não intercept(w0)
    # :normalize -> se normaliza[0,1] X antes da regressão
    model = LinearRegression(fit_intercept=True, normalize=False)
    model.fit(x, y_)

    yp = model.predict(x)

    # scatter dos dados
    ax2 = fig.add_subplot(n, 3, i*3 + 3)
    ax2.scatter(x, y_)
    # plot da função linear
    ax2.plot(x, yp, c='r', linewidth=2.0)
    # coeficientes ficam em .coef_
    # intercept fica em .intercept_
    # cada modelo paramétrico guarda seus parâmetros em atributos terminados em '_'
    ax2.set_title("intercept %f coef %f" % (model.intercept_, model.coef_[0]))
```

From the implementation point of view, this is just plain Ordinary Least Squares (scipy.linalg.lstsq) wrapped as a predictor object.

```python
model = LinearRegression()
model.fit([[0], [1], [2]], [0, 2, 4])
```

### .score(X, y)

Returns the coefficient of determination R^2 of the prediction.

The coefficient R^2 is defined as (1 - u/v), where u is the regression sum of squares ((y_true - y_pred) \** 2).sum() and v is the residual sum of squares ((y_true - y_true.mean()) \** 2).sum(). Best possible score is 1.0, lower values are worse.
