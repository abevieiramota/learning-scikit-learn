```python
# fonte: http://scikit-learn.org/stable/modules/linear_model.html#logistic-regression
# fonte: http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
from sklearn.linear_model import LogisticRegression
import numpy as np
from matplotlib import pyplot as pl
%matplotlib inline
%run util.py
```

```python
classes = [1, 2]
colors = ['r', 'b']
total = 3

fig = pl.figure(figsize=(12, total*4))

for i in xrange(total):

    X0 = np.linspace(-5, 5, 100).reshape((100, 1))
    X1 = np.random.normal(0., 10., 100).reshape((100, 1))
    X = np.hstack((X0, X1))
    Y = np.random.choice(classes, 100)

    i_treino, i_teste = indices([80, 20])
    X_treino, X_teste = X[i_treino, :], X[i_teste, :]
    Y_treino, Y_teste = Y[i_treino], Y[i_teste]

    model = LogisticRegression(penalty='l1')
    model.fit(X_treino, Y_treino)
    Y_ = model.predict(X_teste)
    
    # dados
    ax0 = fig.add_subplot(total, 3, i*3 + 1)
    ax0.set_title(u'dados')
    # classificação correta
    ax1 = fig.add_subplot(total, 3, i*3 + 2)
    ax1.set_title(u'classificação real')
    # classificação do modelo
    ax2 = fig.add_subplot(total, 3, i*3 + 3)
    ax2.set_title(u'classificação do modelo')

    for classe, c in zip(classes, colors):
        ax0.scatter(X[Y==classe][:, 0], X[Y==classe][:, 1], c=c)
        ax1.scatter(X_teste[Y_teste==classe][:, 0], X_teste[Y_teste==classe][:, 1], c=c)
        ax2.scatter(X_teste[Y_==classe][:, 0], X_teste[Y_==classe][:, 1], c=c)
```

```python
X0 = np.linspace(0, 10, 100).reshape((100, 1))
X1 = np.random.normal(0., 20, 100).reshape((100, 1))
X = np.hstack((X0, X1))
Y = np.ones(100)
Y[10] = 0

fig = pl.figure(figsize=(12, 4))
ax0 = fig.add_subplot(1, 3, 1)

ax0.scatter(X0[Y==1], X1[Y==1], c='b')
ax0.scatter(X0[Y==0], X1[Y==0], c='r')
ax0.set_title(u'classificação real')

model_l2 = LogisticRegression(penalty='l2', C=.00001)
model_l2.fit(X, Y)
Y_l2 = model_l2.predict(X)
ax1 = fig.add_subplot(1, 3, 2)
ax1.scatter(X0[Y_l2==1], X1[Y_l2==1], c='b')
ax1.scatter(X0[Y_l2==0], X1[Y_l2==0], c='r')
ax1.set_title(u'penalty l2 e C=.00001')

model_l1 = LogisticRegression(penalty='l1', C=10000.)
model_l1.fit(X, Y)
Y_l1 = model_l1.predict(X)
ax2 = fig.add_subplot(1, 3, 3)
ax2.scatter(X0[Y_l1==1], X1[Y_l1==1], c='b')
ax2.scatter(X0[Y_l1==0], X1[Y_l1==0], c='r')
ax2.set_title(u'penalty l1 e C=10000.')
```

    C : float, optional (default=1.0)
    Inverse of regularization strength; must be a positive float. Like in support vector machines, smaller values specify stronger regularization.
