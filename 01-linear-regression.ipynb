```python
# fonte: http://scikit-learn.org/stable/modules/linear_model.html#ordinary-least-squares
# fonte: http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
from sklearn.linear_model import LinearRegression
import numpy as np
from matplotlib import pyplot as pl
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100).reshape((100, 1))
y = x*2
noises = [(0., .5), (0., 1.), (0., 2.), (.2, .5), (.2, 1.), (.2, 2.)]
n = len(noises)

fig = pl.figure(figsize=(12, n*4))

for i, noise in enumerate(noises):
    
    mu = noise[0]
    sigma = noise[1]

    noise = np.random.normal(mu, sigma, 100).reshape((100, 1))

    y_ = y + noise

    # sem noise
    ax1 = fig.add_subplot(n, 3, i*3 + 1)
    ax1.scatter(x, y)
    ax1.set_title("sem noise")

    # com noise
    ax2 = fig.add_subplot(n, 3, i*3 + 2)
    ax2.scatter(x, y_)
    ax2.set_title("noise: $\mu=%.2f$ $\sigma=%.2f$" % (mu, sigma))

    # linear model
    model = LinearRegression()
    model.fit(x, y_)

    yp = model.predict(x)

    # scatter dos dados
    ax2 = fig.add_subplot(n, 3, i*3 + 3)
    ax2.scatter(x, y_)
    # plot da função linear
    ax2.plot(x, yp, c='r', linewidth=2.0)
    ax2.set_title("linear model. coefs = "+",".join([str(c) for c in model.coef_]))
```
